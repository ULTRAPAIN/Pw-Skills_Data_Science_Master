Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework written in Python that is designed to be simple, lightweight, and flexible. It is built on top of the Werkzeug WSGI (Web Server Gateway Interface) toolkit and the Jinja2 templating engine, both of which are also written in Python. Flask is widely used for building web applications, APIs, and other types of web services.

Advantages of Flask Framework:

Lightweight: Flask is a lightweight framework with minimal dependencies, making it easy to install, use, and maintain.

Flexibility: Flask is highly flexible and can be used for building a wide range of applications, from simple web pages to complex web applications.

Easy to Learn: Flask has a simple and intuitive API that is easy to learn for developers who are new to web development or to Python.

Extensible: Flask provides a range of extensions and plug-ins that make it easy to add new features and functionality to your web application.

Fast: Flask is fast and efficient, with low overhead and minimal resource consumption.

Large Community: Flask has a large and active community of developers who contribute to its development, provide support, and create new extensions and plug-ins.

Testing: Flask comes with built-in support for unit testing, making it easy to write and run tests for your web application.

Overall, Flask is a powerful and flexible web framework that is well-suited for building a wide range of web applications, from simple prototypes to complex, high-performance web services. Its lightweight, easy-to-learn, and extensible design makes it a popular choice among Python developers.

Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of defining the URLs that your application will respond to, and the functions that will handle those requests. App routes are used to map URLs to specific functions in your Flask application.

There are many reasons why we use app routes in Flask:

Handling User Requests: App routes allow you to define the URLs that your application will respond to, and the functions that will handle those requests. This allows you to create dynamic web pages and respond to user input.

Separation of Concerns: App routes help you separate the logic of your application into smaller, more manageable functions. Each function can handle a specific URL, which makes it easier to maintain and update your code.

URL Generation: App routes also allow you to generate URLs for your application dynamically. This is useful when you need to create links between different parts of your application.

Custom Error Handling: You can use app routes to define custom error handlers for your application. For example, you can create a route that handles 404 errors, and return a custom error message or redirect the user to a different page.

Overall, app routing is an essential feature of Flask that allows you to define the URLs that your application will respond to, and the functions that will handle those requests. Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

The url_for() function is used in Flask to generate URLs for specific routes defined in the application. It takes the name of the function as its first argument and any number of keyword arguments, where each keyword corresponds to a URL parameter defined in the route.

Here is an example of using url_for() function in Flask:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to the home page!'

@app.route('/user/<username>')
def user_profile(username):
    return f'Welcome to {username}\'s profile page!'

if __name__ == '__main__':
    with app.test_request_context():
        # generating URLs using url_for function
        print(url_for('home'))
        # Output: /
        print(url_for('user_profile', username='john'))
        # Output: /user/john
```
