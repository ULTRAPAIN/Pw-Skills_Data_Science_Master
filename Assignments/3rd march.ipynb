Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

There are many plots which comes under seaborn 
1:- scatterplot()
2:-displot()
3:-jointplot()
4:-catplot()
5:-pairplot()

1:- scatterplot is the plot which comes in both seaborn as well as matplotlib .this plot techniques marks down by points ,according to the point of x and y axis.
2:-displot also known as dsitribtion plot , this plot is equivalent to histogram plot in matplotlib , this plot is used for distributing the range values which are occuring repeatedly and according to the frequency they match the values.
3:-jointplot combines the plot of scatterplot and histogram(displot) where displot indicates for its one own axis and scatterplot shows relationships between the x and y - axis.
4:- catplot is the plot which is used for categorization of the data .
5:-pairplot is a technique which is used for finding the trends of numerical cloumns.


Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
```

```python
fmri=sns.load_dataset("fmri")
fmri
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,style="event",hue="region")
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,style="region",hue="event")
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
```

```python
titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
sns.boxplot(data=titanic,x="class",y="age")
```

```python
sns.boxplot(data=titanic,x="class",y="fare")
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
```

```python
flight=sns.load_dataset("flights")
flight.head()
```

```python

flights_pivot = flight.pivot("month", "year", "passengers")
flights_pivot.head()
```

```python
sns.heatmap(flights_pivot,fmt="d",annot=True,cmap="YlGnBu")
```
