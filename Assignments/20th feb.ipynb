1:-The GET method is used to retrieve information from a server. When a client sends a GET request, it is asking the server to send back a representation of a resource, which could be a web page, an image, a file, or any other type of data.

2:-GET requests are typically used when a user clicks on a link or enters a URL in a web browser, or when a search engine indexes a web page.

3:-GET requests are considered "safe" and "idempotent," which means that they should not change the state of the server and can be repeated multiple times without affecting the outcome.

POST Method:
    
1:-The POST method is used to submit data to a server for processing. When a client sends a POST request, it is asking the server to accept the data enclosed in the request body and perform an action based on that data, such as saving it to a database, updating a record, or creating a new resource.

2:-POST requests are typically used when a user submits a form on a website, when an application sends data to a server for processing, or when an API (Application Programming Interface) is used to send data between different systems.

3:-POST requests are not considered "safe" or "idempotent" because they can change the state of the server, and repeating the same POST request multiple times may result in different outcomes.

Q3. Why is redirect() used in Flask?

The redirect() function in Flask is used to redirect users from one URL to another. It is a convenient way to redirect users to a different route or external URL within a Flask web application.

There are several reasons why redirect() is commonly used in Flask applications:

Route redirection: redirect() can be used to redirect users from one route to another within the same Flask application. This can be useful in cases where a user requests a certain URL, but the application needs to direct them to a different URL or route based on certain conditions or logic.

External URL redirection: redirect() can also be used to redirect users to external URLs outside of the Flask application. This can be useful when the application needs to redirect users to a third-party website, such as after authentication or when implementing external OAuth authentication.

User experience: Redirecting users to different routes or URLs can help improve the overall user experience of a Flask application. For example, after a user submits a form, the application can use redirect() to redirect them to a confirmation page or a success page, providing feedback and a better user experience.

Q4. What are templates in Flask? Why is the render_template() function used?

In Flask, templates are files that contain the HTML code for rendering web pages. Templates allow developers to separate the presentation logic from the application logic, making it easier to maintain and update the user interface. Flask uses a built-in templating engine called Jinja2, which provides features such as template inheritance, control structures (e.g., loops and conditionals), and variable substitution.

The render_template() function in Flask is used to render templates and generate HTML pages dynamically. It takes the name of the template file as an argument and returns the rendered HTML content that can be sent as a response to the client.
